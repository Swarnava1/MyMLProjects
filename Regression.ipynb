```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# 1. Load the Excel data
df = pd.read_excel('house_data1.xlsx')
df
```

```python
print(df.columns)
```

```python
# 2. Prepare the data
# We reshape X because the model expects a 2D array
X = df[['Square_Feet (X)']]
y = df['Price (y)']

# 3. Initialize and Train the model
model = LinearRegression()
model.fit(X,y)

# 4. Make a prediction
new_house = pd.DataFrame({'Square_Feet (X)': [1800]})
prediction = model.predict(new_house)

new_house
```

```python
prediction
```

```python
sq_ft_value = new_house.iloc[0, 0]
print(f"The predicted price for an {sq_ft_value} sq ft house is: ${prediction[0]:,.2f}")
```

```python
from sklearn.metrics import mean_absolute_error, r2_score

predictions = model.predict(X)

#Calculate MAE
mae = mean_absolute_error(y, predictions)
print(f"Mean Absolute Error (MAE): ${mae:,.2f}")

#Calculate R-Squared
r2 = r2_score(y, predictions)
print(f"R-Squared (R2) Score: {r2:.4f}")
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, r2_score

# 1. Generate 100 random houses (Square feet between 800 and 4000)
np.random.seed(42) # This ensures we both get the exact same random numbers
X_large = np.random.randint(800, 4000, size=(100, 1))

# Generate realistic prices: $150 per sq ft, plus a $50k base, plus some random "noise." 
# (because some houses are nicer than others!)
y_large = (X_large[:, 0] * 150) + 50000 + np.random.normal(0, 30000, 100)

X_large
```

```python
y_large
```

```python
# 2. THE MAGIC SPLIT: 80% for training, 20% for testing
X_train, X_test, y_train, y_test = train_test_split(X_large, y_large, test_size=0.20, random_state=42)

print(f"Training on {len(X_train)} houses...")
print(f"Testing on {len(X_test)} hidden houses...\n")

# 3. Train the model ONLY on the Training data
model_real = LinearRegression()
model_real.fit(X_train, y_train)

# 4. Test the model ONLY on the hidden Testing data
test_predictions = model_real.predict(X_test)

# 5. Grade the real-world performance
real_mae = mean_absolute_error(y_test, test_predictions)
real_r2 = r2_score(y_test, test_predictions)

print(f"REAL-WORLD MAE: ${real_mae:,.2f}")
print(f"REAL-WORLD R-Squared: {real_r2:.4f}")
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, r2_score

# 1. Create a realistic multi-column dataset (100 houses)
np.random.seed(42)
sq_ft = np.random.randint(800, 4000, 100)
bedrooms = np.random.randint(1, 6, 100)
age = np.random.randint(0, 50, 100)

# We secretly generate the price using logic: $150/sqft + $20k/bedroom - $1k/year of age + Base + Noise
price = (sq_ft * 150) + (bedrooms * 20000) - (age * 1000) + 50000 + np.random.normal(0, 15000, 100)

# Put it into a Pandas DataFrame (This is exactly what pd.read_excel would do!)
df_multi = pd.DataFrame({
    'Square_Feet': sq_ft,
    'Bedrooms': bedrooms,
    'Age_Years': age,
    'Price': price
})

df_multi
```

```python
# 2. Prepare X (Features) and y (Target)
# Notice the double brackets [[ ]] - we are passing a list of multiple columns!
X = df_multi[['Square_Feet', 'Bedrooms', 'Age_Years']]
y = df_multi['Price']

# 3. Split the data (80% Train, 20% Test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)

# 4. Train the model
model_multi = LinearRegression()
model_multi.fit(X_train, y_train)

# 5. Evaluate the model on the hidden Test data
predictions = model_multi.predict(X_test)
print("--- MODEL SCORES ---")
print(f"MULTI-FEATURE MAE: ${mean_absolute_error(y_test, predictions):,.2f}")
print(f"MULTI-FEATURE R-Squared: {r2_score(y_test, predictions):.4f}\n")

# 6. THE MAGIC: Let's peek inside the model's brain!
print("--- HOW THE MODEL VALUES A HOUSE ---")
print(f"Base Price (Intercept): ${model_multi.intercept_:,.2f}")

# model.coef_ holds the "weights" the model learned for each column
for feature, weight in zip(X.columns, model_multi.coef_):
    print(f"Every 1 unit of '{feature}' changes the price by: ${weight:,.2f}")
```

```python
# 1. Create a DataFrame for the new house (must match our exact column names)
my_new_house = pd.DataFrame({
    'Square_Feet': [2500],
    'Bedrooms': [4],
    'Age_Years': [10]
})

# 2. Ask the model to predict
estimated_price = model_multi.predict(my_new_house)

print(f"The market value right now for this house is: ${estimated_price[0]:,.2f}")
```
